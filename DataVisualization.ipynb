```python
import numpy as np
import matplotlib.pyplot as plt
```

### Line Plot

###### Basic Line Plot

```python
x= np.linspace(1,10, 100)
y=np.sin(x)

plt.xlabel("x")
plt.ylabel("Sin(x)")
plt.title("Sin Graph")

plt.plot(x,y)
plt.show()
```

###### Custom Line graph

```python
x= np.linspace(1,10, 50)
y=np.sin(x)

plt.xlabel("x")
plt.ylabel("Sin(x)")
plt.title("Sin Graph")

# x-axis, y-axis, style, size, corresponding style, color, opacity
plt.plot(x,y,'o', markersize='5', linestyle="--", color='orange', alpha=.8 )
plt.show()
```

###### Multiple line graphs

```python
x= np.linspace(1,10, 50)
y1=np.sin(x)
y2=np.cos(x)

# x-axis, y-axis, style, size, corresponding style, color, opacity
plt.plot(x,y1,'o', markersize='5', linestyle="--", color='red', alpha=.6 ,label='sinx' )
plt.plot(x,y2,'o', markersize='5', linestyle="--", color='blue', alpha=.6 ,label='cosx')
plt.xlabel("x")
plt.ylabel("Sin(x) & cos(x)")
plt.title("Sin-Cos Graph")
plt.legend()

plt.show()
```

### Scatter Plot

###### Basic Scatter Plot

```python
a=np.random.rand(300)
b=np.random.rand(300)

plt.scatter(a,b)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Scatter Plot")
plt.show()
```

###### Customized Scatter Plot

```python
a=np.random.rand(500)
b=np.random.rand(500)
z=np.random.rand(500)
plt.scatter(a,b,c=z, marker='o', cmap='viridis_r', alpha =0.8)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Scatter Plot")
plt.colorbar()
plt.show()
```
