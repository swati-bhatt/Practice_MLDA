```python
import numpy as np
import matplotlib.pyplot as plt
```

### Line Plot

###### Basic Line Plot

```python
x= np.linspace(1,10, 100)
y=np.sin(x)

plt.xlabel("x")
plt.ylabel("Sin(x)")
plt.title("Sin Graph")

plt.plot(x,y)
plt.show()
```

###### Custom Line graph

```python
x= np.linspace(1,10, 50)
y=np.sin(x)

plt.xlabel("x")
plt.ylabel("Sin(x)")
plt.title("Sin Graph")

# x-axis, y-axis, style, size, corresponding style, color, opacity
plt.plot(x,y,'o', markersize='5', linestyle="--", color='orange', alpha=.8 )
plt.show()
```

###### Multiple line graphs

```python
x= np.linspace(1,10, 50)
y1=np.sin(x)
y2=np.cos(x)

# x-axis, y-axis, style, size, corresponding style, color, opacity
plt.plot(x,y1,'o', markersize='5', linestyle="--", color='red', alpha=.6 ,label='sinx' )
plt.plot(x,y2,'o', markersize='5', linestyle="--", color='blue', alpha=.6 ,label='cosx')
plt.xlabel("x")
plt.ylabel("Sin(x) & cos(x)")
plt.title("Sin-Cos Graph")
plt.legend()

plt.show()
```

### Scatter Plot

###### Basic Scatter Plot

```python
a=np.random.rand(500)
b=np.random.rand(500)

plt.scatter(a,b)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Scatter Plot")
plt.show()
```

###### Customized Scatter Plot

```python
a=np.random.rand(500)
b=np.random.rand(500)
z=np.random.rand(500)
plt.scatter(a,b,c=z, marker='o', cmap='viridis_r', alpha =0.8)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Customized Scatter Plot")
plt.colorbar()
plt.show()
```

###### Scatter Plot with multiple Data sets

```python
x1=np.random.rand(300)
y1=np.random.rand(300)
x2=np.random.rand(300)
y2=np.random.rand(300)
x3=np.random.rand(300)
y3=np.random.rand(300)

plt.scatter(x1, y1, marker='o', alpha =0.8, label='x1')
plt.scatter(x2, y2, marker='o', alpha =0.8, label='x2')
plt.scatter(x3, y3, marker='o', alpha =0.8, label='x3')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Scatter Plot")
plt.legend()
plt.show()
```

###### Hexbining Scatter Plot

```python
a=np.random.rand(10000)
b=np.random.rand(10000)

plt.hexbin(a,b, gridsize=20)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Hexbining Scatter Plot")
plt.colorbar()
plt.show()
```

### Bar Graph

###### Basic Bar Graph

```python
x= np.array(["A", "B", "C", "D"])
y= np.array([10,20,30, 40])

plt.bar(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Basic Bar Graph")

plt.show()
```

###### Stacked Bar Graph

```python
data= np.random.rand(10,5)

# print(data)

# plt.bar(np.arange(10), data[:,0]+ data[:,1]+ data[:,2]+ data[:,3]+ data[:,4] ,label='1')
plt.bar(np.arange(10), data[:,0], label=1)
plt.bar(np.arange(10), data[:,1], bottom= data[:,0], label=2)
plt.bar(np.arange(10), data[:,2], bottom= data[:,0]+ data[:,1], label=3)
plt.bar(np.arange(10), data[:,3], bottom= data[:,0]+ data[:,1]+ data[:,2], label=4)
plt.bar(np.arange(10), data[:,4], bottom= data[:,0]+ data[:,1]+ data[:,2] +data[:,3], label=5)


plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Stacked Bar Graph")
plt.legend()

plt.show()
```

###### Grouped Bar Graph

```python
data= np.array([[10, 20, 30],
               [40,50, 60],
               [70, 80, 90]])
# print(data)

x=np.arange(3)

width=0.25

plt.bar(x, data[0,:], width, label='1', color="red", alpha=0.9)
plt.bar(x+width, data[1,:], width, label='2' , alpha=0.9)
plt.bar(x+2*width, data[2,:], width, label='3', color="green", alpha=0.9)

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Grouped Bar Graph")
plt.legend()

plt.show()
```

### Histogram

###### Basic Histogram

```python
data=np.random.randn(1000000)

plt.hist(data, bins=1000)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Histogram")

plt.show()
```

###### Cumulative Histogram

```python
data=np.random.randn(1000000)

plt.hist(data, bins=1000, cumulative=True)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Cumulative Histogram")

plt.show()
```

###### Histogram with Multiple DataSets

```python
data1=np.random.randn(10000000)
data2=np.random.randn(10000000)+3

plt.hist(data1, bins=1000, alpha=0.7, label='1')
plt.hist(data2, bins=1000, alpha=0.7, label='2')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Histogram")
plt.legend()
plt.show()
```

### Box Plot

```python
data= np.random.rand(100,5)

plt.boxplot(data, labels=['BP1', 'BP2', 'BP3', 'BP4', 'BP5'], vert=True)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Box Plot")

plt.show()
```

### HeatMap

```python
data= np.random.rand(10,10)

plt.imshow(data, cmap='viridis')
plt.colorbar()
plt.show()
```

### Pie Chart

```python
labels = ['Python', 'C++', 'Java', 'JavaScript', 'PHP']
sizes =  [21.3,        14.6 , 9.2,     8.5,       7.7]
explode = (0.08,        0,    0,       0,           0)
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightpink']

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.2f%%', shadow=True, startangle=300,)
plt.axis('equal')
plt.title('Programmming Languages')
plt.legend(title="Languages", loc='upper left')
plt.show()
```
